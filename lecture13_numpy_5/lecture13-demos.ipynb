```python
arr = np.random.rand(50,10)
to_add = np.arange(10)
print(arr)
```

```python
print(to_add)
```

```python
arr[0]
```

```python
# first row should be
arr[0] + to_add
```

```python
# we could write a for loop
result = np.zeros((50,10))
for ii in range(50):
    result[ii] = arr[ii] + to_add

# this suuuucks
```

```python
print(result)
```
