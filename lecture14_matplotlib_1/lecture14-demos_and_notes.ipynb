```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
```

```python
np.set_printoptions(precision=2) # only print 2 decimal places
```

```python
x = np.linspace(0, 10, 100)
print(x)
```

```python
y = np.sin(x)
print(y)
```

```python
plt.plot(x, y)
```

```python
x2 = np.linspace(0, 10, 15)
y2 = np.sin(x2)

plt.plot(x2, y2)
```

```python
# just plot dots
plt.plot(x, y, '.')
```

```python
# plot big dots with a line
plt.plot(x, y, 'o-')
```

```python
# plot squares connected by a dotted line
plt.plot(x2, y2, 's:')
```

```python
# alternate version? NO
plt.plot(x, y, '-.')
```

```python
# alternate order!
plt.plot(x, y, '-o')
```

```python
plt.plot(x, y, '--')
```

```python
y_fast = np.sin(x*3)
```

```python
plt.plot(x, y)
plt.plot(x, y_fast)
```

```python
# plt.figure() # HIDDEN!
plt.plot(x, y)

plt.figure()
plt.plot(x, y_fast)
```

```python
# the non-*sigh*-MATLAB way to do matplotlib

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x, y)
ax.plot(x, y_fast)
```

```python
xs = [1, 3, 5, 7, 9]
ys = [14, 22, -6, 12]

plt.plot(xs, ys)
```
