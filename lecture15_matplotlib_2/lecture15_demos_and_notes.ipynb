## imshow & matshow

```python
arr = np.zeros((20, 20))
arr[:,1::2] = np.arange(20)[:,np.newaxis]
print(arr)
```

```python
plt.matshow(arr)
```

```python
arr2 = arr.copy()
arr2[4,12] = 10000

plt.matshow(arr2)
```

```python
plt.matshow(arr2)
plt.colorbar()
```

```python
plt.matshow(arr2, vmin=0, vmax=25)
plt.colorbar()
```

```python
plt.matshow(arr2, vmin=5, vmax=25)
plt.colorbar()
```

```python
plt.matshow(arr2, vmin=0, vmax=25, cmap='Reds')
plt.colorbar()
```

```python
plt.matshow(arr2, vmin=0, vmax=25, cmap=plt.cm.Reds)
plt.colorbar()
```
