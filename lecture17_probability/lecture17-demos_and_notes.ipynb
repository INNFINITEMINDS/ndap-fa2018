```python
# coin simulation!
coin_flips = np.random.rand(10000, 100) < 0.5
print(coin_flips.shape)
```

```python
coin_flips.sum()
```

```python
coin_flip_heads = coin_flips.sum(axis=1)
print(coin_flip_heads.shape)
```

```python
plt.plot(coin_flip_heads)
```

```python
plt.hist(coin_flip_heads, bins=15)
```

```python
coin_flip_heads.max()
```

```python
coin_flip_heads.min()
```

```python
# compute the p-value for our outcome, 63 heads
num_extreme = ((coin_flip_heads >= 63).sum() + (coin_flip_heads <= (100 - 63)).sum())
print(num_extreme)
```

```python
p_value = num_extreme / 10000
print(p_value)
```
