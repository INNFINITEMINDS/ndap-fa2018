```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
q = 0.5

coin_flips = np.random.rand(100000, 50) < q
print(coin_flips.shape)
```

```python
num_heads = coin_flips.sum(axis=1)
print(num_heads.shape)
```

```python
plt.hist(num_heads, np.arange(51));
```

```python
import scipy.misc
```

```python
ks = np.arange(51)
```

```python
c = scipy.misc.comb(50, ks)
```

```python
plt.plot(c)
```

```python
p1 = q ** ks
print(p1)
```

```python
p2 = (1 - q) ** (50 - ks)
print(p2)
```

```python
binom = c * p1 * p2 # Pr(k heads in n trials)
```

```python
plt.hist(num_heads, np.arange(51) - 0.5, density=True);
plt.plot(ks, binom, 'k-')
```
