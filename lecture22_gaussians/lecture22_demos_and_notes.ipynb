```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
```

```python
rands = np.random.rand(100, 10000)
rand_means = rands.mean(axis=0)
rand_means.shape
```

```python
plt.hist(rands.ravel(), 100);
```

```python
plt.hist(rand_means, 50);
```

```python
# let's get weird with it
rands2 = np.random.rand(100, 100000) ** 3
rand_means2 = rands2.mean(0)
```

```python
plt.hist(rands2.ravel(), 100);
```

```python
plt.hist(rand_means2, 50);
```

```python
# let's look at sample variance
```

```python
gauss_rands = np.random.randn(20, 1000000)
```

```python
real_variance = ((gauss_rands - gauss_rands.mean(axis=0))**2).sum(axis=0) / 20
real_variance2 = gauss_rands.var(axis=0)
```

```python
np.allclose(real_variance, real_variance2)
```

```python
plt.hist(real_variance, 100);
```

```python
real_variance.mean()
```

```python
# compute sample variance
sample_variance = ((gauss_rands - gauss_rands.mean(axis=0))**2).sum(axis=0) / (20 - 1)
sample_variance2 = gauss_rands.var(axis=0, ddof=1)
```

```python
plt.hist(sample_variance, 100);
```

```python
sample_variance.mean()
```

```python
sample_variance2.mean()
```
