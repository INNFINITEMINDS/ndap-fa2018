```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
```

```python
from scipy.stats import norm
```

```python
x = np.linspace(-4, 4, 200)
```

```python
plt.plot(x, norm.pdf(x))
```

```python
plt.plot(x, norm.cdf(x))
```

```python
plt.plot(x, norm.sf(x))
plt.plot(x, 1 - norm.cdf(x))

print(norm.sf(0.9))
```

```python
y = np.linspace(0, 1, 200)

plt.plot(y, norm.isf(y))
```

```python
norm.isf(0.9) # what is the value of x such that norm.sf(x) = 0.9
# 90% of the values we draw will be > -1.28...
```

```python
# standard errors
arr = 0.5*np.random.randn(50) + 1
print(arr[:5])
```

```python
sample_mean = arr.mean()
print(sample_mean)
```

```python
# get the bootstrap distribution of the mean
bootstrap_means = np.array([np.random.choice(arr, 50).mean() for _ in range(1000)])
```

```python
plt.hist(bootstrap_means, 20);
```

```python
bootstrap_sem = bootstrap_means.std()
print(bootstrap_sem)
```

```python
analytic_sem = arr.std(ddof=1) / np.sqrt(50)
print(analytic_sem)
```

```python
# z-test
```

```python
mean = 0.5
var = 1.7

x = 2.4
```

```python
# what is the probability of observing a value greater than 2.4 from a gaussian distribution
# with mean 0.5 and variance 1.7?
```

```python
norm.sf(0.5) # the Pr[x > 0.5] in standard normal
```

```python
# compute a z-score!
z_x = (x - mean) / np.sqrt(var)
print(z_x)
```

```python
# what is the prob. of observing a value greater than 1.457... from a standard normal?
```

```python
norm.sf(z_x)
```
