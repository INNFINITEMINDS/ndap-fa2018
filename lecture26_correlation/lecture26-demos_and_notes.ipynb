```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
```

```python
arr1 = np.random.randn(50)
print(arr1)
```

```python
# how do we make another array that is correlated with arr1, but not perfectly correlated?
```

```python
# multiply by a scalar?
arr2 = arr1 * 1.5
```

```python
np.corrcoef(arr1, arr2)
# no, this makes something that's perfectly correlated (correlation is 1.0)
```

```python
plt.plot(arr1, arr2, '.')
```

```python
# we could make an uncorrelated array
arr3 = np.random.randn(50)
```

```python
np.corrcoef(arr1, arr3)

# but then the correlation is random
```

```python
plt.plot(arr1, arr3, '.')
```

```python
# create array that is correlated but not identical to arr1
# by adding some noise to arr1!
arr4 = arr1 + np.random.randn(50)
```

```python
# there we go
np.corrcoef(arr1, arr4)
```

```python
plt.plot(arr1, arr4, '.')
```

```python
plt.plot(arr1, '.-')
plt.plot(arr4, '.-')
```
